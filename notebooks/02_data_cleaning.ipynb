```python
import pandas as pd

df = pd.read_excel("../data/Online_Retail.xlsx")

print(df.shape)
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
# Suppression des valeurs nulles importantes
df = df.dropna(subset=['CustomerID'])

print(df.shape)
```

```python
# Suppression des doublons
df = df.drop_duplicates()

print(df.shape)
```

```python
# Gestion des retours (Quantité négatives)
df = df[df['Quantity'] > 0]

print(df.shape)
```

```python
# Suppression des prix nuls ou négatifs
df = df[df['UnitPrice'] > 0]

print(df.shape)
```

```python
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate']) # Conversion des dates
```
